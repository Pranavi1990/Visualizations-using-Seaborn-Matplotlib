```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel('week 18.xlsx')
```

## 1.    Delete the Unnamed 0 column

```python
df=df.drop(columns='Unnamed: 0')
```

```python
df.columns
```

## Plot the total_bill column histogram

```python
sns.histplot(df['total_bill'])
plt.show()
```

### 3.    Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.figure(figsize=(5,5),dpi=100)
ax=sns.scatterplot(x='total_bill',y='tip',data=df,palette='pastel',color='orange')
ax.set_title('Scatter plot btwn total_bill and tip')
plt.show()
```

```python
plt.figure(figsize=(5,5),dpi=100)
ax=sns.regplot(x='total_bill',y='tip',data=df,color='red')
ax.set_title('Scatter plot btwn total_bill and tip with regplot()')
plt.show()
```

```python
plt.figure(figsize=(6,5),dpi=100)
ax=sns.lmplot(x='total_bill',y='tip',data=df,ci=None)
ax.fig.subplots_adjust(top=0.9)
ax.fig.suptitle('Scatter plot btwn total_bill and tip with lmplot()')
plt.show()
```

### 4.    Create one image with the relationship of total_bill, tip and size.

```python

ax=sns.relplot(
    data=df, x="total_bill", y="tip", size="size",hue="size",sizes=(25,200))
ax.fig.suptitle("Scatter plot showing relationship between total_bill,tip and size")
plt.show()
```

### 5.    Present the relationship between days and total_bill value

```python
plt.figure(figsize=(7,5),dpi=100)
ax=sns.lineplot(x='day',y='total_bill',ci=None,data=df)
ax.set_title('Relationship between days and total_bill amount')
plt.show()
```

### 6.    Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.scatterplot(x='tip',y='day',hue='sex',data=df)
```

### 7.    Create a box plot presenting the total_bill per day differentiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(5,5),dpi=100)
sns.boxplot(x='day',y='total_bill',hue='time',data=df)
plt.show()
```

### 8.    Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
sns.displot(data=df, x="tip", col="time")
plt.show()
```

### 9.    Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker

```python
sns.lmplot(x='total_bill',y='tip',data=(df),col='sex',hue='smoker',ci=None)
plt.show()
```
